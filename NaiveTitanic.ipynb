```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('titanic.xls')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
missing_age=df.loc[df['age'].isnull()]
```

```python
missing_age.head()
```

```python
df.groupby(['survived','sex','pclass'])['age'].mean()
```

```python
df['age'].fillna(df.groupby(['survived','sex','pclass'])['age'].transform('mean'),inplace=True)
```

```python
mals=list(missing_age.index)
```

```python
df.iloc[mals].head()
```

```python
df.isnull().sum()
```

```python
embark=df.loc[df['embarked'].isnull()]
```

```python
embark
```

```python
embarks=list(embark.index)
```

```python
df['embarked'].value_counts()
```

```python
df['embarked'].fillna('S',inplace=True)
```

```python
df.iloc[embarks]
```

```python
df.isnull().sum()
```

```python
df = pd.get_dummies(data=df, columns=['pclass','embarked'])
df.head()
```

```python
df['family_num'] = df['sibsp'] + df['parch']
df.drop(['sibsp', 'parch'], axis=1, inplace=True)
df.head()
```

```python
df['TravelAlone']=np.where((df['family_num'] > 0), 0, 1)
df.head()
```

```python
df['sex']=df['sex'].map({'female':0,'male':1})
```

```python
df.head()
```

# Gaussian Nave Bayes

```python
from sklearn.naive_bayes import GaussianNB   #import Gaussian Bayes modeling function
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
df.dtypes
```

```python
modeldf = df[['survived','age', 'sex', 'TravelAlone']]
```

```python
modeldf.corr()
```

```python
X = modeldf.drop('survived', axis=1)

#column of predictive target values
y = modeldf['survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=109)
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_train, y_train)
```

```python
gnb.score(X_train, y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
cm = pd.DataFrame(
    confusion_matrix(y_test, y_pred),
    columns=['Predicted Failed', 'Predicted Passed'],
    index=['True Failed', 'True Passed']
)

cm
```
